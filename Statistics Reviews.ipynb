```python
import pandas as pd
from tqdm import tqdm
```

# Camera

```python
camera=pd.read_table('D:/Eigene Dateien/Desktop/Amazon_data/amazon_reviews_us_Camera_v1_00.tsv', error_bad_lines=False)
camera.head()
```

```python
camera.groupby("star_rating").count()
```

```python
camera['review_date']=pd.to_datetime(camera['review_date'], errors='coerce')
```

```python
camera_grouped=camera.groupby(camera.review_date.dt.year).count()
camera_grouped = camera_grouped.drop(columns=['review_date'])
camera_grouped = camera_grouped.reset_index()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

fig, ax = plt.subplots()

for axis in [ax.xaxis, ax.yaxis]:
    axis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.title("camera")
plt.xlabel('year')
plt.ylabel('#ratings')   
plt.plot(camera_grouped['review_date'], camera_grouped['star_rating'])
plt.show()
```
